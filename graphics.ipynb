```python
import pandas as pd
import matplotlib.pyplot as plt

log_file = "logs/training_log_v1.csv"

try:
    data = pd.read_csv(log_file)
    print(f"Archivo '{log_file}' cargado correctamente. Total de {len(data)} Ã©pocas registradas.")

    # --- GRÃFICA 1: PÃRDIDAS (LOSSES) ---
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'].to_numpy(), data['Generator_Loss'].to_numpy(), label="PÃ©rdida del Generador (v1)", color="blue")
    plt.plot(data["Epoch"].to_numpy(), data["Discriminator_Loss"].to_numpy(), label="PÃ©rdida del Discriminador (v1)", color="orange")
    plt.title("EvoluciÃ³n de las pÃ©rdidas durante el entrenamiento (v1)")
    plt.xlabel("Ãpoca")
    plt.ylabel("PÃ©rdida (Loss)")
    plt.legend()
    plt.grid(True)
    # He corregido una pequeÃ±a errata en la ruta para guardar, para que coincida con la de la otra grÃ¡fica
    plt.savefig("results/v1/grafica_perdidas_v1.png")
    plt.show()
    print("GrÃ¡fica de pÃ©rdidas guardada como 'grafica_perdidas_v1.png'")

    # --- GRÃFICA 2: PRECISIÃN (ACCURACY) ---
    plt.figure(figsize=(12, 6))
    # AÃADIMOS .to_numpy() a cada columna de pandas
    plt.plot(data['Epoch'].to_numpy(), (data['Discriminator_Accuracy'] * 100).to_numpy(), label='PrecisiÃ³n del Discriminador (v1)', color="green")
    plt.title("EvoluciÃ³n de la precisiÃ³n del Discriminador (v1)")
    plt.xlabel("Ãpoca")
    plt.ylabel("PrecisiÃ³n (%)")
    plt.axhline(y=75, color="r", linestyle='--', label="Equilibrio Ideal (75%)")
    
    plt.legend()
    plt.grid(True)
    plt.savefig("results/v1/grafica_precision_v1.png")
    plt.show()
    print("GrÃ¡fica de precisiÃ³n guardada como 'grafica_precision_v1.png'")

except FileNotFoundError:
    print(f"Error: No se encontrÃ³ el archivo '{log_file}'. AsegÃºrate de que estÃ¡ en la misma carpeta que el notebook")
except Exception as e:
    print(f"OcurriÃ³ un error inesperado: {e}")
```
